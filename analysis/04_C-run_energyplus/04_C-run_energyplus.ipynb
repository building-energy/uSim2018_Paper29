# 04_C-run_energyplus

This notebook reads the modified gbXML files, runs an EnergyPlus simulation and saves the results.

```python
import sys,os
fp=os.path.join(os.pardir)
if not fp in sys.path: sys.path.append(fp)
import openbuilding as ob
```

```python
for i in range(0,360,5):
    
    print(i)
    
    #reads the gbXML file
    fn='detached_house_{}.gbxml'.format(i)
    fp=r'../03_D-new_gbXML/'+fn
    gbxml=ob.GbxmlGraph()
    gbxml.read_xml(fp)
    
    #sets up and runs an EnergyPlus model
    ep=ob.EnergyPlusModel()
    ep.input_gbxml=gbxml
    ep.input_epw=r'GBR_Birmingham.035340_IWEC.epw'
    ep.simulation_folder=r'../05_D-energyplus_results/sim-detached_house_{}'.format(i)
    ep.run_gbxml()
    
    #saves the results as a BimGraph and a graphml file    
    bim=ep.output_bim
    fn='detached_house_{}.pickle'.format(i)
    fp=r'../05_D-energyplus_results/'+fn
    bim.write_pickle(fp)
    fn='detached_house_{}.graphml'.format(i)
    fp=r'../05_D-energyplus_results/'+fn
    bim.write_graphml(fp)
```
